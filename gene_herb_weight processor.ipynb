```python
import pandas as pd
import os

path = r'C:\tg'  # directory where SMHBXXXXX.csv exist

for i, tang in enumerate([T1, T2, T3, T4, T5, T6, T7, T8, T9, T10, T11, T12, T13]):
    weights = pathway_weights[i]
    data_list = []
    tang_name = tang[0]
    print(tang_name)

    for herb in tang[1:]:
        print(herb)
        file_path = os.path.join(path, herb + '.csv')
        if not os.path.exists(file_path):
            print(f"NO {file_path} file")
            continue

        df = pd.read_csv(file_path)

        # check there is not numerical value 
        df = df[pd.to_numeric(df['P_value'], errors='coerce').notna()]
        df = df[pd.to_numeric(df['Value'], errors='coerce').notna()]

        # use whole df since GSEA analysis use whole gene set 
        filtered_df = df

        # add to new data frame
        for _, row in filtered_df.iterrows():
            gene = row['Gene symbol']
            value = float(row['Value'])
            score = value * weights.get(herb, 1) #/ sum(weights.values())  # 가중치 적용
            data_list.append([herb, gene, score])

    # create final dataframe
    output_df = pd.DataFrame(data_list, columns=['herb', 'GeneSymbol', 'Score'])

    # saving
    output_file = os.path.join(path, tang_name + '_processed.csv')
    output_df.to_csv(output_file, index=False)#, encoding='utf-8')
    print(f"{output_file} save complete!")
```
